```python
import mysqk.connector
import os
import time

def MinPrice():
    
    #Connecting to the Database
    connection = mysql.connector.connect(host='localhost', # Setting up the database connection
                                         database = 'beeronomics',
                                         user='root',
                                         password='Cheers!!')

    #Creates a Cursor (called cursor1) that connects to the database
    cursor1 = connection.cursor(buffered = True)
    cursor2 = connection.cursor(buffered = True)
    cursor3 = connection.cursor(buffered = True)
    cursor4 = connection.cursor(buffered = True)
    
    #mySQL Query Variables
    mySql_Select_Beer_Ids = """SELECT beer_id FROM beer_types"""
    mySql_Select_Beer_Price_Act = """SELECT beer_price_active FROM beer_types"""
    mySql_Select_Beer_Price_Min = """SELECT beer_price_min FROM beer_types"""
    mySql_Select_Start_Price = """SELECT beer_price_start FROM beer_types"""
    
    #Executing the Queries and Fetching the Data for Python
    records1 = cursor1.execute(mySql_Select_Beer_Ids)
    results1 = cursor1.fetchall()
    records2 = cursor2.execute(mySql_Select_Beer_Price_Act)
    results2 = cursor2.fetchall()
    records3 = cursor3.execute(mySql_Select_Beer_Price_Min)
    results3 = cursor3.fetchall()
    records4 = cursor4.execute(mySql_Select_Start_Price)
    results4 = cursor4.fetchall()
    
    # Row Count Function
    rows = cursor1.rowcount
    
    #Creating Empty Lists
    beer_ids = []
    beer_prices_act = []
    beer_prices_min = []
    beer_prices_start = []
    
    #Converting Tuples From mySQL to Lists
    for t in results1: 
        for x in t:
            beer_ids.append(x)
    print(beer_ids)
    
    for t in results2: 
        for x in t:
            beer_prices_act.append(x)
    print(beer_prices_act)
    
    for t in results3: 
        for x in t:
            beer_prices_min.append(x)
    print(beer_prices_min)
    
    for t in results4: 
        for x in t:
            beer_prices_start.append(x)
    print(beer_prices_start)
    
    
    # Loop to Update Values to the Min Price
    for i in range(0,rows):
        beer_prices_act[i] = beer_prices_min[i]
        print(beer_prices_act[i])

    # mySQl Update Query Variable
    mySql_Update_Beer_Prices = """UPDATE beer_types 
                                SET beer_price_active = %s 
                                WHERE beer_id = %s"""
    
    # Turning New Price List into Tuple
    beer_prices_new = (*beer_prices_act,)
    
    # Cursor for Price Reassignment
    cursor = connection.cursor(buffered = True)
    
    # Updating Active Prices
    for i in range(0,rows):
        update = cursor.execute(mySql_Update_Beer_Prices,(beer_prices_new[i],beer_ids[i],))
        connection.commit()
    
    # Closing Database Connections
    cursor1.close()
    cursor2.close()
    cursor3.close()
    cursor4.close()
    connection.close()

def StartPrice():
    
    #Connecting to the Database
    connection = mysql.connector.connect(host='localhost', # Setting up the database connection
                                         database = 'beeronomics',
                                         user='root',
                                         password='Cheers!!')

    #Creates a Cursor (called cursor1) that connects to the database
    cursor1 = connection.cursor(buffered = True)
    cursor2 = connection.cursor(buffered = True)
    cursor3 = connection.cursor(buffered = True)
    cursor4 = connection.cursor(buffered = True)
    
    #mySQL Query Variables
    mySql_Select_Beer_Ids = """SELECT beer_id FROM beer_types"""
    mySql_Select_Beer_Price_Act = """SELECT beer_price_active FROM beer_types"""
    mySql_Select_Beer_Price_Min = """SELECT beer_price_min FROM beer_types"""
    mySql_Select_Start_Price = """SELECT beer_price_start FROM beer_types"""
    
    #Executing the Queries and Fetching the Data for Python
    records1 = cursor1.execute(mySql_Select_Beer_Ids)
    results1 = cursor1.fetchall()
    records2 = cursor2.execute(mySql_Select_Beer_Price_Act)
    results2 = cursor2.fetchall()
    records3 = cursor3.execute(mySql_Select_Beer_Price_Min)
    results3 = cursor3.fetchall()
    records4 = cursor4.execute(mySql_Select_Start_Price)
    results4 = cursor4.fetchall()
    
    # Row Count Function
    rows = cursor1.rowcount
    
    #Creating Empty Lists
    beer_ids = []
    beer_prices_act = []
    beer_prices_min = []
    beer_prices_start = []
    
    #Converting Tuples From mySQL to Lists
    for t in results1: 
        for x in t:
            beer_ids.append(x)
    print(beer_ids)
    
    for t in results2: 
        for x in t:
            beer_prices_act.append(x)
    print(beer_prices_act)
    
    for t in results3: 
        for x in t:
            beer_prices_min.append(x)
    print(beer_prices_min)
    
    for t in results4: 
        for x in t:
            beer_prices_start.append(x)
    print(beer_prices_start)
    
    
    # Loop to Update Values to the Min Price
    for i in range(0,rows):
        beer_prices_act[i] = beer_prices_start[i]
        print(beer_prices_start[i])

    # mySQl Update Query Variable
    mySql_Update_Beer_Prices = """UPDATE beer_types 
                                SET beer_price_active = %s 
                                WHERE beer_id = %s"""
    
    # Turning New Price List into Tuple
    beer_prices_new = (*beer_prices_act,)
    
    # Cursor for Price Reassignment
    cursor = connection.cursor(buffered = True)
    
    # Updating Active Prices
    for i in range(0,rows):
        update = cursor.execute(mySql_Update_Beer_Prices,(beer_prices_new[i],beer_ids[i],))
        connection.commit()
    
    # Closing Database Connections
    cursor1.close()
    cursor2.close()
    cursor3.close()
    cursor4.close()
    connection.close()
    
def Timer():
    s=0
    m=0
    while s <= 60:
        os.system('cls')
        #print (m, 'Minutes', s, 'Seconds')
        time.sleep(1)
        s += 1
        if s == 60:
            m += 1
            s = 0
            print("One minute has been reached.")
    if m == 5:
        break

Id = bool(input("Would you like to start the program?"))

if id == 1:
    
    Timer()
    
    if m == 1:
        MinPrice()
        
    if m == 2:
        StartPrice()
else:
    
    print("Why?")
```

```python
# Timer
import time
print "               TIMER"
#Ask for Duration
Dur2 = input("How many minutes?: ")
Dur3 = input("How many seconds?: ")
TDur = Dur2 * 60 + Dur3
# Ask to Begin
start = raw_input("Would you like to begin Timing? (y/n): ")
if start == "y":
    timeLoop = True

# Variables to keep track and display
CSec = 0
Sec = 0
Min = 0
Hour = 0
# Begin Process
timeLoop = start
while timeLoop:
    CSec += 1
    Sec += 1
    print(str(Hour) + " Hours " + str(Min) + " Mins " + str(Sec) + " Sec ")
    time.sleep(1)
    if Sec == 60:
        Sec = 0
        Min += 1
        Hour = 0
        print(str(Min) + " Minute(s)")
    if Min == 60:
        Sec = 0
        Min = 0
        Hour += 1
        print(str(Hour) + " Hour(s)")
    elif CSec == TDur:
        timeLoop = False
        print("time\'s up")
        input("")
```
