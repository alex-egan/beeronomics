```python
import mysql.connector
from tkmacosx import Button
from tkinter import *

main = Toplevel()
main.title('Beeronomics Main Menu')
main.configure(bg='black')
main.attributes("-fullscreen",True)
maintitle_lbl = Label(main, text = "Beeronomics")
maintitle_lbl.grid()


##Connecting to the Database
connection = mysql.connector.connect(host='localhost', # Setting up the database connection
                                     database = 'beeronomics',
                                     user='root',
                                     password='Cheers!!')

c1 = connection.cursor(buffered = True)
c2 = connection.cursor(buffered = True)
mySql_Select_Query1 = """SELECT beer_name FROM beer_types"""
mySql_Select_Query2 = """SELECT beer_price_active FROM beer_types"""
c1.execute(mySql_Select_Query1)
c2.execute(mySql_Select_Query2)
beer_name = c1.fetchall()
beer_price_act = c2.fetchall()
rows = c1.rowcount
c1.close()
c2.close()
beer_names = []
beer_prices_act = []
output = []

for t in beer_name:
    for x in t:
        beer_names.append(x)

for t in beer_price_act:
    for x in t:
        beer_prices_act.append(x)

for t in range(0,rows):
    output.append(t)
    
def CreateNewBeer():
    
    def ExecuteCreation():
        connection = mysql.connector.connect(host='localhost', # Setting up the database connection
                                             database = 'beeronomics',
                                             user='root',
                                             password='Cheers!!')
        cursor = connection.cursor(buffered = True)
        mySql_Insert_Query = """INSERT INTO beer_types (beer_name, beer_price_min, beer_price_max, beer_price_active, purchase_count, beer_price_start) 
                              VALUES (%s,%s,%s,%s,%s,%s)"""
        beer_name1 = beer_name.get()
        print(beer_name1)
        beer_price_min1 = float(beer_minprice.get())
        beer_price_max1 = float(beer_maxprice.get())
        beer_price_active1 = float(beer_actprice.get())
        purchase_count1 = beer_purcount.get()
        beer_price_start1 = float(beer_srtprice.get())
        
        t = ()
        
        t = t + (beer_name1,) + (beer_price_min1,) + (beer_price_max1,) + (beer_price_active1,) + (purchase_count1,) + (beer_price_start1,)
        
        cursor.executemany(mySql_Insert_Query,(t,))
        connection.commit()
        
        success_lbl = Label(CreateBeerScreen, text = beer_name1 + " was added successfully.")
        success_lbl.grid()
        
        cursor.close()
        
    def ResetTextBoxes():
        beer_name.delete(0, END)
        beer_minprice.delete(0, END)
        beer_maxprice.delete(0, END)
        beer_actprice.delete(0, END)
        beer_purcount.delete(0, END)
        beer_srtprice.delete(0, END)
        success_lbl.destroy()
        
    def Return():
        CreateBeerScreen.destroy()
        Tk.update(self)
        
    
    CreateBeerScreen = Toplevel()
    CreateBeerScreen.title("Create New Beer in Database")
    beer_name = Entry(CreateBeerScreen, width=30)
    beer_name.grid(row=0,column=1,columnspan=2)
    beer_minprice = Entry(CreateBeerScreen, width=30)
    beer_minprice.grid(row=1,column=1,columnspan=2)
    beer_maxprice = Entry(CreateBeerScreen, width=30)
    beer_maxprice.grid(row=2,column=1,columnspan=2)
    beer_actprice = Entry(CreateBeerScreen, width=30)
    beer_actprice.grid(row=3,column=1,columnspan=2)
    beer_purcount = Entry(CreateBeerScreen, width=30)
    beer_purcount.grid(row=4,column=1,columnspan=2)
    beer_srtprice = Entry(CreateBeerScreen, width=30)
    beer_srtprice.grid(row=5,column=1,columnspan=2)
    
    beer_name_lbl = Label(CreateBeerScreen, text = "Beer Name", width = 15, anchor = W)
    beer_name_lbl.grid(row=0,column=0)
    beer_minprice_lbl = Label(CreateBeerScreen, text = "Beer Minimum Price", width=15, anchor = W)
    beer_minprice_lbl.grid(row=1,column=0)
    beer_maxprice_lbl = Label(CreateBeerScreen, text = "Beer Maximum Price", width=15, anchor = W)
    beer_maxprice_lbl.grid(row=2,column=0)
    beer_actprice_lbl = Label(CreateBeerScreen, text = "Beer Active Price", width=15, anchor = W)
    beer_actprice_lbl.grid(row=3,column=0)
    beer_purcount_lbl = Label(CreateBeerScreen, text = "Purchase Count", width=15, anchor = W)
    beer_purcount_lbl.grid(row=4,column=0)
    beer_srtprice_lbl = Label(CreateBeerScreen, text = "Start Price", width=15, anchor = W)
    beer_srtprice_lbl.grid(row=5,column=0)
    
    create_beer_btn = Button(CreateBeerScreen, width=5, height=1, text = "Add Entry", command = ExecuteCreation)
    create_beer_btn.grid(row=6,column=1,columnspan=1)
    
    exit_create_beer_btn = Button(CreateBeerScreen, width=10, height=1, text = "Back to Home", command = Return)
    exit_create_beer_btn.grid(row=6,column=2,columnspan=1)
    
    reset_text_boxes_btn = Button(CreateBeerScreen, width=10, height=1, text = "Reset Text Boxes", command = ResetTextBoxes)
    reset_text_boxes_btn.grid(row=6,column=0,columnspan=1)
    
j = 0
k = 0

for i in range(0,rows):
    output[i] = Button(main, wraplength = 60, highlightbackground = 'black', text=beer_names[i], width=5, height=5, command = lambda: PriceCall())
    output[i].grid(row=j, column=k)
    j += 1
    if j == 4:
        k += 1
        j = 0
print(j)
        
create_beer_button = Button(main, text = "Create New Beer in Database", command = CreateNewBeer)
create_beer_button.grid(row=0,column=k+1)

root = Tk()
root.title('Login')
login_lbl = Label(root, text = "Login", width = 15, anchor = W)
login_lbl.grid(row=0,column=1)

def Login():
    
    un = username.get()
    pw = password.get()
    
    uname = 'beeronomics'
    pword = 'Cheers!!'
    
    if un == uname and pw == pword:
        main.mainloop()
    
    else:
        wrong_login_lbl = Label(root, "Credentials were incorrect.")
        wrong_login_lbl.grid(row=4,column=1)
        

username = Entry(root, width=30)
username.grid(row=1,column=1,columnspan=2)
password = Entry(root, width=30)
username.grid(row=2,column=1,columnspan=2)
username_lbl = Label(root, text = "Username", width=15, anchor = W)
username_lbl.grid(row=1,column=0)
password_lbl = Label(root, text = "Password", width=15, anchor = W)
password_lbl.grid(row=2,column=0)
login_button = Button(root, text = "Login", command = Login)
login_button.grid(row=3,column=1)

root.mainloop()
```
