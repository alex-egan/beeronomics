```python
## Getting a Button to Do Something

from tkinter import *

root = Tk()

def myClick():
    myLabel = Label(root,text="Look, I clicked the Button!")
    myLabel.pack()

## When using command for a button, don't use the parentheses like you would
## for a normal function
myButton = Button(root, text = "Click Me!", command=myClick)
myButton.pack()

root.mainloop()
```

```python
## Creating a Simple Calculator App
from tkinter import *

root = Tk()

##Creating a Page Title
root.title("Simple Calculator")

## Entry Widget (Input)
e = Entry(root,width=35,borderwidth=5)
e.grid(row=0,column=0,columnspan=4,padx=10,pady=10)

#e.insert(0,)

def button_click(number):
    current = e.get()
    e.delete(0, END)
    e.insert(0, str(current) + str(number))
    return

def button_clear():
    e.delete(0, END)

def button_equal():
    second_number = int(e.get())
    e.delete(0, END)
    
    if math == "addition":
        e.insert(0, f_num + second_number)
        
    if math == "subtraction":
        e.insert(0, f_num - second_number)
        
    if math == "multiplication":
        e.insert(0, f_num * second_number)
        
    if math == "division":
        e.insert(0, f_num / second_number)
    
def button_add():
    first_number = int(e.get())
    global f_num
    global math
    math = "addition"
    f_num = first_number
    e.delete(0, END)

def button_subtract():
    first_number = int(e.get())
    global f_num
    global math
    math = "subtraction"
    f_num = first_number
    e.delete(0, END)

def button_multiply():
    first_number = int(e.get())
    global f_num
    global math
    math = "multiplication"
    f_num = first_number
    e.delete(0, END)

def button_divide():
    first_number = int(e.get())
    global f_num
    global math
    math = "division"
    f_num = first_number
    e.delete(0, END)

# Define the Buttons

button_1 = Button(root,text="1",padx=40,pady=20,command=lambda: button_click(1))
button_2 = Button(root,text="2",padx=40,pady=20,command=lambda: button_click(2))
button_3 = Button(root,text="3",padx=40,pady=20,command=lambda: button_click(3))
button_4 = Button(root,text="4",padx=40,pady=20,command=lambda: button_click(4))
button_5 = Button(root,text="5",padx=40,pady=20,command=lambda: button_click(5))
button_6 = Button(root,text="6",padx=40,pady=20,command=lambda: button_click(6))
button_7 = Button(root,text="7",padx=40,pady=20,command=lambda: button_click(7))
button_8 = Button(root,text="8",padx=40,pady=20,command=lambda: button_click(8))
button_9 = Button(root,text="9",padx=40,pady=20,command=lambda: button_click(9))
button_0 = Button(root,text="0",padx=40,pady=20,command=lambda: button_click(0))
button_add = Button(root,text="+",padx=39,pady=20,command=button_add)
button_subtract = Button(root,text="-",padx=39,pady=20,command=button_subtract)
button_multiply = Button(root,text="*",padx=39,pady=20,command=button_multiply)
button_divide = Button(root,text="/",padx=39,pady=20,command=button_divide)
button_equal = Button(root,text="=",padx=39,pady=20,command=button_equal)
button_clear = Button(root,text="Clear",padx=32,pady=20,command=button_clear)

# Put the buttons on the screen

button_1.grid(row=3,column=0)
button_2.grid(row=3,column=1)
button_3.grid(row=3,column=2)

button_4.grid(row=2,column=0)
button_5.grid(row=2,column=1)
button_6.grid(row=2,column=2)

button_7.grid(row=1,column=0)
button_8.grid(row=1,column=1)
button_9.grid(row=1,column=2)

button_0.grid(row=4,column=0)

button_clear.grid(row=4,column=2)
button_add.grid(row=1,column=3)
button_equal.grid(row=4,column=1)
button_subtract.grid(row=2,column=3)
button_multiply.grid(row=3,column=3)
button_divide.grid(row=4,column=3)

root.mainloop()
```

```python
## Creating Radio Buttons

from tkinter import *
from PIL import ImageTk,Image

root = Tk()
root.title('learn to code at codemy.com')

## Defining a Kinter Variable
r = IntVar() ##or StrVar() for strings
r.set("2")

MODES = [
    ("Pepperoni","Pepperoni"),
    ("Cheese","Cheese"),
    ("Mushroom","Mushroom"),
    ("Onion","Onion"),
]

pizza = StringVar()
pizza.set("Pepperoni")

for text,mode in MODES:
    Radiobutton(root, text=text, variable=pizza, value=mode).pack(anchor=W)

def clicked(value):
    myLabel = Label(root, text=value)
    myLabel.pack()

myLabel = Label(root, text=pizza.get())
myLabel.pack()
##Creating Radio Buttons
Button(root,text="Option 1",variable=r, value=1, command=lambda: clicked(r.get())).pack()
Button(root,text="Option 2",variable=r, value=2, command=lambda: clicked(r.get())).pack()

myButton = Button(root, text="Click Me!", command=lambda: clicked(pizza.get()))
myButton.pack()

root.mainloop()
```

```python
## Creating a Table

import mysql.connector

## Connect to DB
connection = mysql.connector.connect(host='localhost', # Setting up the database connection
                                     database = 'beeronomics',
                                     user='root',
                                     password='Cheers!!')

#Create Cursor
cursor1 = connection.cursor(buffered = True)

#Create Table
mySql_Create_Table = """Create Table addresses (
                first_name text,
                last_name text,
                address text,
                city text,
                state text,
                zipcode int

)"""
cursor1.execute(mySql_Create_Table)

#Commit Changes
connection.commit()

#Close Connection
cursor1.close()
connection.close()
```

```python
from tkinter import *
from PIL import ImageTk,Image
import mysql.connector

root = Tk()
root.title('Address Book')

#Create Submit Function for Database

def submit():
    conn = mysql.connector.connect(host='localhost', # Setting up the database connection
                                database = 'beeronomics',
                                user='root',
                                password='Cheers!!')

    c = conn.cursor(buffered = True)
    
    c.execute("""INSERT INTO addresses VALUES (:f_name, :l_name, :address, :city, :state, :zipcode)""",
            {
                 'f_name':f_name.get(),
                 'l_name':l_name.get(),
                 'address':address.get(),
                 'city':city.get(),
                 'state':state.get(),
                 'zipcode':zipcode.get()
            })
    #Clear the Text Boxes
    f_name.delete(0, END)
    l_name.delete(0, END)
    address.delete(0, END)
    city.delete(0, END)
    state.delete(0, END)
    zipcode.delete(0, END)
    
    conn.commit()

    conn.close()

f_name = Entry(root, width=30)
f_name.grid(row=0,column=1,padx=20)
l_name = Entry(root, width=30)
l_name.grid(row=1,column=1)
address = Entry(root, width=30)
address.grid(row=2,column=1)
city = Entry(root, width=30)
city.grid(row=3,column=1)
state = Entry(root, width=30)
state.grid(row=4,column=1)
zipcode = Entry(root, width=30)
zipcode.grid(row=5,column=1)

f_name_label = Label(root, text="First Name")
f_name_label.grid(row=0,column=0)
l_name_label = Label(root, text="Last Name")
l_name_label.grid(row=1,column=0)
address_label = Label(root, text="Address")
address_label.grid(row=2,column=0)
city_label = Label(root, text="City")
city_label.grid(row=3,column=0)
state_label = Label(root, text="State")
state_label.grid(row=4,column=0)
zipcode_label = Label(root, text="Zipcode")
zipcode_label.grid(row=5,column=0)

#Create Submit Button
submit_btn = Button(root, text="Add Record to Database", command=submit)
submit_btn.grid(row=6,column=0,columnspan=2,pady=10,padx=10,ipadx=100)

root.mainloop()
```
